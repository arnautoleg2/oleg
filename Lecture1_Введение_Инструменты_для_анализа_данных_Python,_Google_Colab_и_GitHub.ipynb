## Python example (https://python-graph-gallery.com/341-python-gapminder-animation/)

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv')
data['continent'] = pd.Categorical(data['continent'])
data.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

# Subset of the data for year 1952
data1952 = data[data.year == 1952]

# Scatterplot
ax.scatter(
    x = data1952['lifeExp'],
    y = data1952['gdpPercap'],
    s = data1952['pop']/50000,
    c = data1952['continent'].cat.codes,
    cmap = "Accent",
    alpha = 0.6,
    edgecolors = "white",
    linewidth = 2
)

# Add titles (main and on axis)
plt.yscale('log')
ax.set_xlabel("Life Expectancy")
ax.set_ylabel("GDP per Capita")
ax.set_title("Year 1952")
ax.set_ylim(100,50000)
ax.set_xlim(30, 75)

plt.show()
```

#3. Google Colab

- **Что такое Среда разработки?**

Среда разработки (Development Environment) — это программа, которая помогает программистам создавать, отлаживать и поддерживать код. Обычно она включает в себя редактор кода, компилятор или интерпретатор, отладчик, а также другие инструменты для работы с кодом.

- **Что такое Google Colab?**
  
Это облачная среда разработки (Сloud Development Environment, CDE), то есть работает в облачной среде и доступна через браузер, а не устанавливается на локальном компьютере программиста. Это означает, что аналитик может разрабатывать и отлаживать код где угодно и с любого устройства, имеющего доступ к интернету.

- **Преимущества перед локальными средами разработки (Integrated Development Environment, IDE)**.
